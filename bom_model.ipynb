```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = np.load('/Users/cristiancruz/Library/CloudStorage/GoogleDrive-cristianluke2k@gmail.com/My Drive/REL C 200/REL C 200 Project/bom_embeddings.npy')
df = pd.read_csv('/Users/cristiancruz/Library/CloudStorage/GoogleDrive-cristianluke2k@gmail.com/My Drive/REL C 200/REL C 200 Project/bom_data_cleaned.csv')
df.head()
```

```python
verses = df['scripture_text'].tolist()
```

## Testing it out

```python
def get_similar_verses(input, model, embeddings, num_to_return:int=5, print_result=False):
  from sklearn.metrics.pairwise import cosine_similarity
  import numpy as np

  query_embedding = model.encode([input])

  similarities = cosine_similarity(query_embedding, embeddings)[0]

  top_indices = similarities.argsort()[-num_to_return:][::-1]

  if print_result:
    print(f'Printing the top {num_to_return} results...')
    for idx in top_indices:
      verse = df.iloc[idx]
      print(f"{verse['verse_title']}: {verse['scripture_text']}")

  return top_indices
```

```python
input = "I don't know how to feel the Spirit."

results = get_similar_verses(input=input,model=model, embeddings=embeddings, print_result=True)
```
